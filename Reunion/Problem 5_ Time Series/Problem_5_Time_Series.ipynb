```python
import pandas as pd
```

```python
file_path = '/content/Time Series.xlsx'
```

```python
df = pd.read_excel(file_path)
```

```python
df['Start'] = pd.to_datetime(df['Start'])
df['End'] = pd.to_datetime(df['End'])
```

```python
df = df.sort_values(by=['Name', 'Start']).reset_index(drop=True)
```

```python
def merge_periods(group):
    merged_data = []
    current_start = group.iloc[0]['Start']
    current_end = group.iloc[0]['End']
    current_activities = [group.iloc[0]['Activity']]

    for i in range(1, len(group)):
        if group.iloc[i]['Start'] <= current_end:
            current_end = max(current_end, group.iloc[i]['End'])
            current_activities.append(group.iloc[i]['Activity'])
        else:
            merged_data.append({
                'Name': group.iloc[i-1]['Name'],
                'Start': current_start,
                'End': current_end,
                'Activities': current_activities
            })
            current_start = group.iloc[i]['Start']
            current_end = group.iloc[i]['End']
            current_activities = [group.iloc[i]['Activity']]

    merged_data.append({
        'Name': group.iloc[len(group)-1]['Name'],
        'Start': current_start,
        'End': current_end,
        'Activities': current_activities
    })

    return pd.DataFrame(merged_data)

merged_df = df.groupby('Name').apply(lambda x: merge_periods(x)).reset_index(drop=True)
```
